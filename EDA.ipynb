# 2D Scatter plot
It visualize the relationship between two variables, and to identify any patterns or trends in the data.  By plotting data points in two dimensions, a scatter plot can reveal the overall distribution of the data, as well as any clusters, outliers, or other structures in the data.

```python
sns.set_style('whitegrid')
sns.scatterplot(x='SepalLengthCm', y = 'SepalWidthCm', hue='Species', data=iris)
plt.show()
```

From above shown 2D scatter plot respect to SepalLengthCm vs SeaplWidthCm,  we can easily determined that there is a distinct difference between iris setosa and the rest two types of flower.

## pair plot
A pair plot is a type of visualization that displays pairwise relationships between variables in a dataset. It is also sometimes referred to as a scatter plot matrix. In a pair plot, the diagonal shows the distribution of each variable in the dataset, while the off-diagonal plots show the relationship between each pair of variables.
Pair plots are useful for identifying potential relationships and patterns between variables, and for identifying any outliers or unusual observations in the data. They can also be used to compare the distribution of different variables, and to identify any potential areas for further analysis or modeling.

```python
# pair plot
iris = iris.drop('Id', axis=1)
plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris, hue="Species", height=3)
plt.show()
```

From pair plot we can see that we have most differentiable distribution with respect to petalwidth and then petallength.

# Histogram, PDF, CDF
A histogram is a graphical representation of the distribution of a dataset. It consists of a set of rectangles or bars, where the area of each bar corresponds to the frequency or relative frequency of the data in each interval or bin. Histograms are commonly used to visualize the shape, center, and spread of a dataset, as well as any potential outliers or unusual observations.

Probability Density Function (PDF):
The probability density function (PDF) is a mathematical function that describes the probability distribution of a continuous random variable. The PDF provides the probability of a variable falling within a specific range of values. It is calculated by taking the derivative of the cumulative distribution function (CDF) of the random variable. The area under the PDF curve between two values represents the probability of the variable falling between those two values.

Cumulative Distribution Function (CDF):
The cumulative distribution function (CDF) is a mathematical function that describes the probability that a random variable takes on a value less than or equal to a certain value. It is defined as the integral of the probability density function (PDF) over the range of possible values. The CDF is a monotonically increasing function, and ranges from 0 to 1. The CDF can be used to determine probabilities for a continuous random variable, by calculating the difference between the CDF at two different values.

```python
sns.histplot(iris, x='PetalLengthCm', hue='Species', kde=True)
plt.show()
```

```python
sns.histplot(iris, x = 'PetalWidthCm', hue='Species', kde= True)
plt.show()
```

```python
sns.histplot(iris, x = 'SepalWidthCm', hue='Species', kde='True')
plt.show()
```

```python
iris_setosa = iris.loc[iris["Species"] == "Iris-setosa"]
iris_virginica = iris.loc[iris["Species"] == "Iris-virginica"]
iris_versicolor = iris.loc[iris["Species"] == "Iris-versicolor"]

counts, bin_edges = np.histogram(iris_setosa['PetalLengthCm'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(iris_setosa['PetalLengthCm'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();
```

# Mean, variance and standard deviation

```python
print("Means: ")
print(np.mean(iris_setosa["PetalLengthCm"]))
print(np.mean(iris_virginica["PetalLengthCm"]))
print(np.mean(iris_versicolor["PetalLengthCm"]))

print("Std-dev: ")
print(np.std(iris_setosa["PetalLengthCm"]))
print(np.std(iris_virginica["PetalLengthCm"]))
print(np.std(iris_versicolor["PetalLengthCm"]))
```

```python
# Median, Quantiles, Percentiles, IQR
print("Medians: ")
print(np.median(iris_setosa["PetalLengthCm"]))
print(np.median(iris_virginica["PetalLengthCm"]))
print(np.median(iris_versicolor["PetalLengthCm"]))

print("\nQuantiles: ")
print(np.percentile(iris_setosa["PetalLengthCm"], np.arange(0, 100, 25)))
print(np.percentile(iris_virginica["PetalLengthCm"], np.arange(0, 100, 25)))
print(np.percentile(iris_versicolor["PetalLengthCm"], np.arange(0, 100, 25)))
```

# Box plot and whiskers
A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the distribution of the data and any potential outliers. The plot consists of a box, which represents the interquartile range (IQR) of the data, and whiskers, which extend from the box to show the range of the data.

The box in the plot represents the middle 50% of the data, with the bottom and top edges of the box representing the first and third quartiles of the data, respectively. The horizontal line inside the box represents the median value of the data.

The whiskers in the plot extend from the box to the minimum and maximum values of the data, or to a maximum of 1.5 times the IQR from the edges of the box. Any data points outside of the whiskers are considered potential outliers and are plotted as individual points.

Box plots are useful for visualizing the spread and skewness of a dataset, as well as identifying any potential outliers. They are commonly used in data analysis and statistical applications.

```python
# Boxplot
sns.boxplot(iris, x ='Species', y='PetalLengthCm')
plt.show()
```

## Violin plot
A violin plot is a type of graphical representation of a dataset that combines features of a box plot and a kernel density plot. It is used to show the distribution of the data and to compare multiple distributions.

In a violin plot, the shape of the plot represents the density of the data at different values. The wider parts of the plot represent areas where there is a higher density of data, while the narrower parts represent areas with lower density.

The plot is split in half, with each half representing a different group or variable being compared. The violin plots for each group are mirrored, so that they are symmetrical around the vertical axis.

The central line in the plot represents the median of the data, while the shape and width of the plot show the shape of the distribution. The area inside the plot represents the interquartile range (IQR) of the data.

```python
#Violin plot 
sns.violinplot(iris, x= 'Species', y = 'PetalLengthCm', height= 8)
plt.show()
```

The plot depicts the petal length density and the spread of of iris setosa, versicolor and virginica flower. There is a distinct difference between the iris setosa and others also slight overlap between versicolor and virginica

## Contour plot
contour plot represents the density of points at different locations in the plot. Areas with a higher density of points will be represented by darker colors in the contour plot.

```python
# Multivariate probability density, Contour plot
sns.jointplot(iris_setosa, x ='PetalLengthCm', y = 'PetalWidthCm', kind= 'kde')
plt.show()
```

From the plot we can see that the highest density of iris setosa flower's petal length is around at 1.5cm and the petal width at 0.2 cm.
